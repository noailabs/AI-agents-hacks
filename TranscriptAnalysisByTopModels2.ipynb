```python
from google.colab import userdata
from openai import OpenAI
import pprint
import time

# https://github.com/marketplace/models
client_github = OpenAI(
  base_url= "https://models.inference.ai.azure.com",
  api_key=userdata.get('GITHUB_KEY'),
)

# https://build.nvidia.com/nim
client_nv = OpenAI(
  base_url = "https://integrate.api.nvidia.com/v1",
  api_key = userdata.get('NVIDIA_API_KEY')
)

# https://console.groq.com/docs/models
client_groq = OpenAI(
  base_url="https://api.groq.com/openai/v1",
  api_key=userdata.get('GROQ_API_KEY'),
)


models = [
    [client_github, "gpt-4o"],
    [client_github, "gpt-4o-mini"],
    [client_github,"Mistral-large-2407"],
    [client_nv, "meta/llama-3.1-405b-instruct"],
    [client_groq,"llama-3.1-70b-versatile"]
]


client,model=models[0][0],models[0][1]
temperature=1
top_p=1
max_tokens=4096

def llm(messages):
  print(f"{model},temperature={temperature},top_p={top_p},max_tokens={max_tokens}")
  completion = client.chat.completions.create(
  model=model,
  messages=messages,
  temperature=temperature,
  top_p=top_p,
  max_tokens=max_tokens,
  stream=False
  )
  if not hasattr(completion, 'choices'):
    print(completion)
    raise Exception(completion)

  return completion.choices[0].message.content


def llmp(prompt):
  messages=[{"role": "user","content": prompt}]
  return llm(messages)
```

```python
# Inference test EN
for i in range(len(models)):
  client=models[i][0]
  model=models[i][1]
  resp=llmp("Tell a joke about a horse")
  print(resp)
  time.sleep(1)
  print("")
```
