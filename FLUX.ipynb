```python
## flux
# prompt: a high-fashion woman dancing, wearing sleek black latex and futuristic VR-headset, with dramatic lighting highlighting her sharp features and glossy textures against a deep red background, captured in an editorial, photo-realistic style
https://huggingface.co/spaces/black-forest-labs/FLUX.1-schnell
```

```python
from huggingface_hub import InferenceClient
from google.colab import userdata

client = InferenceClient("black-forest-labs/FLUX.1-schnell", token=userdata.get('HF_TOKEN'))

# output is a PIL.Image object
image = client.text_to_image("Astronaut riding a horse")

image
```
